```python
bs = 50

crop_size = 512

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(size=crop_size, scale=(0.75,1.0)),
    transforms.ColorJitter(contrast=0.5),
    transforms.ColorJitter(brightness=0.85),
    transforms.RandomAffine(0, translate=(0.4, 0.4)),
    transforms.RandomRotation(15),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()
])

test_transform = transforms.Compose([
    transforms.Resize(520),
    transforms.CenterCrop(crop_size),
    transforms.ToTensor()
])

train_set = datasets.ImageFolder("New Masks Dataset/Train", transform=train_transform)
trainloader = DataLoader(train_set, batch_size=bs, shuffle=True, num_workers=4)

test_set = datasets.ImageFolder("New Masks Dataset/Validation", transform=test_transform)
testloader = DataLoader(test_set, batch_size=bs, shuffle=True)

test_set = datasets.ImageFolder("New Masks Dataset/Test", transform=test_transform)
testloader = DataLoader(test_set, batch_size=bs)
```

```python
feature, target = next(iter(trainloader))
feature.shape
```

```python
import torch
from torch import nn, optim
from jcopdl.callback import Callback, set_config

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

```python
from jcopdl.layers import linear_block, conv_block
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            conv_block(3, 4),
            conv_block(4, 8),
            conv_block(8, 16),
            conv_block(16, 32),
            conv_block(32, 64),
            conv_block(64, 128),
            conv_block(128, 256),
            nn.Flatten()
        )
        
        self.fc = nn.Sequential(
            linear_block(4096, 1028, dropout=0.25),
            linear_block(1028, 256, dropout=0.25),
            linear_block(256, 64, dropout=0.25),
            linear_block(64, 16, dropout=0.25),
            linear_block(16, 2, dropout=0.25),            
            linear_block(2, 1, activation="sigmoid")
        )
        
    def forward(self, x):
        return self.fc(self.conv(x))
```

```python
config = set_config({
    "batch_size":bs,
    "crop_size":crop_size,
    "output_activation": "sigmoid",
})
```

```python
model = CNN().to(device)
criterion = nn.BCELoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001)
callback = Callback(model, config, outdir="model")
```

```python
from tqdm.auto import tqdm

def loop_fn(mode, dataset, dataloader, model, criterion, optimizer, device):
    if mode == "train":
        model.train()
    elif mode == "test":
        model.eval()
    cost = correct = 0
    for feature, target in tqdm(dataloader, desc=mode.title()):
        feature, target = feature.to(device), target.to(device)
        output = model(feature)
        loss = criterion(output, target)
        
        if mode == "train":
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        
        cost += loss.item() * feature.shape[0]
        preds = (output > 0.5).to(torch.float32)
        correct += (preds == target).sum().item()
    cost = cost / len(dataset)
    acc = correct / len(dataset)
    return cost, acc
```

```python
while True:
    train_cost, train_score = loop_fn("train", train_set, trainloader, model, criterion, optimizer, device)
    with torch.no_grad():
        test_cost, test_score = loop_fn("test", test_set, testloader, model, criterion, optimizer, device)
    
    # Logging
    callback.log(train_cost, test_cost, train_score, test_score)

    # Checkpoint
    callback.save_checkpoint()
        
    # Runtime Plotting
    callback.cost_runtime_plotting()
    callback.score_runtime_plotting()
    
    # Early Stopping
    if callback.early_stopping(model, monitor="test_score"):
        callback.plot_cost()
        callback.plot_score()
        break
```
